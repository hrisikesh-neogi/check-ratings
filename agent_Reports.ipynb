```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
class agent_Report:
    
    def __init__(self, file_path):
        """
        params:
            file_path = the path of the agents sheet.
        
        """
        self.file_path = file_path
        self.df = pd.read_excel('AgentPerformance.xlsx', skiprows=1) 
        self.df = self.df[self.df['Total Chats'] >0]

        
    def available_Agents(self):
        return list(set(self.df['Agent Name'].to_list()))
        
    def data(self, agent_name):
        self.agent_name = agent_name
        
        
        agent_data = self.df[self.df['Agent Name'] == self.agent_name]
        self.agent_data = agent_data.reset_index(drop=True)
        
        self.total_data = {
            'Agent_name' : agent_name,
            'total_working_Data' : len(self.agent_data[self.agent_data['Total Chats'] > 0]),
        'total_chats' : sum(self.agent_data['Total Chats'].to_list()),
        'min_chat' : min(self.agent_data['Total Chats'].to_list()),
        'max_chat' : max(self.agent_data['Total Chats'].to_list()),
        
        'total_feedback' : sum(self.agent_data['Total Feedback'].to_list()),
        'Avg_rating_of_total_working_days' : sum(self.agent_data['Average Rating'].to_list())/len(self.agent_data[self.agent_data['Total Chats'] > 0]),
        
        'rating < 3.5' : len(self.agent_data[self.agent_data['Average Rating'] < 3.5]),
        
        'rating < 4' : len(self.agent_data[self.agent_data['Average Rating'] < 4]),
        'rating <3' : len(self.agent_data[self.agent_data['Average Rating'] < 3]),
        'rating > 4.5': len(self.agent_data[self.agent_data['Average Rating'] > 4.5])
        }
        
        
        
        
        self.total_data = pd.DataFrame([self.total_data])
        return self.total_data
    
    def final_data(self):
        final_dataset = pd.DataFrame()
        for name in self.available_Agents():
            data = self.data(agent_name=name)
            final_dataset = final_dataset.append(data)
            
        final_dataset = final_dataset.reset_index(drop=True)
            
        
        return final_dataset
        
    
```

```python
agents = agent_Report('AgentPerformance.xlsx')
```
